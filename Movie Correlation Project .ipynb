```python
#This project is about finding correlations for high gross profit movies.

#Begin

# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # This will adjust configuration on plots later on
pd.options.mode.chained_assignment = None


# Read in the data 
df = pd.read_csv('/Users/chrisalvaro/Desktop/Python leraning/movies.csv')
```

```python
df.dropna(inplace=True)
  # Identify missing Data with For loop 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df_numerized = df 

# Use for loop

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Number')

plt.xlabel('Movie Features')

plt.ylabel('Budget for Film')


plt.show()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Number')

plt.xlabel('Movie Features')

plt.ylabel('Budget for Film')


plt.show()
```

```python
##Unstacking this 
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs =  corr_pairs.sort_values()

sorted_pairs.tail(25)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.45]

high_corr.head(10)
```
